```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() 
HAR1 = HAR.apply(le.fit_transform)
```

```python
X = HAR1.drop(['subject', 'Activity','ActivityName'], axis=1)
Y = HAR1['ActivityName']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .20, random_state = 40)
```

# APPLYING RANDOM FOREST CLASSIFIER

```python

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model = rf.fit(X_train,Y_train)
predicted = model.predict(X_test)
```

```python
predicted
```

```python
pd.DataFrame(predicted ,Y_test)
```

```python
from sklearn.metrics import accuracy_score
pred=model.predict(X_train)
accuracy_score(Y_train,pred)
```

```python
pred1=model.predict(X_test)
accuracy_score(Y_test,pred1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,pred1)
```

```python
 from sklearn.metrics import classification_report
print(classification_report(pred1, Y_test))
```

# APPLYING K-NEAREST NEIGHBORS

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors = 24) #setting k= square root of 561
```

```python
knn.fit(X_train,Y_train)
ypred = knn.predict(X_test)
```

```python
print (accuracy_score(Y_test,ypred)*100) 
```

```python
scores = []
for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors = i, n_jobs = -1)
    knn.fit(X_train, Y_train)
    ypred = knn.predict(X_test)
    scores.append(accuracy_score(Y_test,ypred))
```

```python
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy Score')
xticks = range(1,50)
plt.plot(xticks, scores, color='red', linestyle='solid', marker='o',
         markerfacecolor='blue', markersize=5)
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 20})
plt.show()
```

```python
import numpy as np
scores = np.array(scores)
```

```python
#Final result after training the model
print('Optimal No. Of Neighbors: ', scores.argmax()+1)
print('Accuracy Score: '+ str(scores.max()*100)+ ' %')
```
